```python
#Importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score
# load data
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
dfHeart_cleveland = pd.read_csv(url, header=None, na_values=['?'])
# load data
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.hungarian.data'
dfHeart_hungarian = pd.read_csv(url, header=None, na_values=['?'])
# Load the dataset from a CSV file
#dfHeart_cleveland = pd.read_csv("processed.cleveland.csv", header=None, na_values=['?'])
#dfHeart_hungarian = pd.read_csv("processed.hungarian.csv", header=None, na_values=['?'])
# Print the number of rows and columns in the cleveland dataset
print("The number of rows and columns in the cleveland dataset are: ")
print('Number of rows:', dfHeart_cleveland.shape[0])
print('Number of columns:', dfHeart_cleveland.shape[1])

# Print the number of rows and columns in the hungarian dataset
print("The number of rows and columns in the hungarian dataset are:")
print('Number of rows:', dfHeart_hungarian.shape[0])
print('Number of columns:', dfHeart_hungarian.shape[1])
# Print the first 5 rows of the cleveland dataset
print("The first 5 rows of the cleveland dataset are:")
print(dfHeart_cleveland.head())
# Print the first 5 rows of the hungarian dataset
print("\nThe first 5 rows of the hungarian dataset are:")
print(dfHeart_hungarian.head())
# Print the data type of each column in the dataset
print("The data type of each column in the cleveland dataset")
print(dfHeart_cleveland.dtypes)
# Print the data type of each column in the dataset
print("\nThe data type of each column in the hungarian dataset")
print(dfHeart_hungarian.dtypes)
# Print the summary statistics of the cleveland dataset
print("The summary statistics of the cleveland dataset ")
print(dfHeart_cleveland.describe())
# Print the summary statistics of the hungarian dataset
print("\nThe summary statistics of the hungarian dataset ")
print(dfHeart_hungarian.describe())
# Count the number of missing values in each column of cleveland dataset
missing_counts1 = dfHeart_cleveland.isna().sum()
print("Number of missing values in each column of cleveland dataset :\n", missing_counts1)
# Count the number of missing values in each column of hungarian dataset
missing_counts2 = dfHeart_hungarian.isna().sum()
print("\nNumber of missing values in each column of hungarian dataset :\n", missing_counts2)
#Since the missing Values in cleveland dataset are very minimal using below approach of removing
# Remove rows with missing values from cleveland dataset
num_rows_before = dfHeart_cleveland.shape[0]
dfHeart_cleveland = dfHeart_cleveland.dropna()
num_rows_after = dfHeart_cleveland.shape[0]
# Identify rows and columns that were removed
removed_rows = dfHeart_cleveland.loc[~dfHeart_cleveland.index.isin(dfHeart_cleveland.dropna().index)]
removed_cols = []
print("Number of rows before removing missing values in cleveland dataset:", num_rows_before)
print("Number of rows after removing missing values in cleveland dataset:", num_rows_after)
# count number of missing values in each column of cleveland dataset
missing_values_count1 = dfHeart_cleveland.isnull().sum()
# Identify columns that were removed
for col in dfHeart_cleveland.columns:
    if dfHeart_cleveland[col].isnull().sum() == num_rows_after:
        removed_cols.append(col)
print("Number of missing values removed in cleveland dataset: ", np.sum(missing_values_count1))
print("Removed rows with missing values:\n", removed_rows)
print("Removed columns with missing values:\n", removed_cols)
#Since the missing Values in hungarian dataset are more using below approach of removing/replacing using mean and mode imputation
# replace '?' with NaN
dfHeart_cleveland.replace('?', np.nan, inplace=True)
#According to the attribute information for the Heart Disease UCI dataset
#the categorical columns are: sex, cp, fbs, restecg, exang, slope, ca, and thal
#the numerical columns are: age, trestbps, chol, thalach, and oldpeak.
# separate numerical and categorical columns
cat_cols = [1, 2, 5, 6, 8, 10, 11]
num_cols = [0, 3, 4, 7, 9]
# fill missing values in numerical columns with mean
missing_values_count = dfHeart_cleveland[num_cols].isnull().sum()
dfHeart_cleveland[num_cols] = dfHeart_cleveland[num_cols].apply(lambda x: x.fillna(x.mean()))
replaced_mean_count = missing_values_count - dfHeart_cleveland[num_cols].isnull().sum()
print("Number of missing values in numerical columns:", missing_values_count.sum())
print("Number of missing values replaced in numerical columns with mean:", replaced_mean_count.sum())
# fill missing values in categorical columns with mode
missing_values_count = dfHeart_cleveland[cat_cols].isnull().sum()
dfHeart_cleveland[cat_cols] = dfHeart_cleveland[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
replaced_mode_count = missing_values_count - dfHeart_cleveland[cat_cols].isnull().sum()
print("Number of missing values in categorical columns:", missing_values_count.sum())
print("Number of missing values replaced in categorical columns with mode:", replaced_mode_count.sum())
# check if there are still missing values in the dataframe
print(dfHeart_cleveland.isnull().sum())
#Since the missing Values in hungarian dataset are more using below approach of removing/replacing using mean and mode imputation
# replace '?' with NaN
dfHeart_hungarian.replace('?', np.nan, inplace=True)
# separate numerical and categorical columns
#num_cols = [0, 3, 4, 5, 6, 7, 8, 10, 11, 12]
#cat_cols = [1, 2, 9, 13]
cat_cols = [1, 2, 5, 6, 8, 10, 11]
num_cols = [0, 3, 4, 7, 9, 12]
# fill missing values in numerical columns with mean
missing_values_count = dfHeart_hungarian[num_cols].isnull().sum()
dfHeart_hungarian[num_cols] = dfHeart_hungarian[num_cols].apply(lambda x: x.fillna(x.mean()))
replaced_mean_count = missing_values_count - dfHeart_hungarian[num_cols].isnull().sum()
print("Number of missing values in numerical columns:", missing_values_count.sum())
print("Number of missing values replaced in numerical columns with mean:", replaced_mean_count.sum())
# fill missing values in categorical columns with mode
missing_values_count = dfHeart_hungarian[cat_cols].isnull().sum()
dfHeart_hungarian[cat_cols] = dfHeart_hungarian[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
replaced_mode_count = missing_values_count - dfHeart_hungarian[cat_cols].isnull().sum()
print("Number of missing values in categorical columns:", missing_values_count.sum())
print("Number of missing values replaced in categorical columns with mode:", replaced_mode_count.sum())
# check if there are still missing values in the dataframe
print(dfHeart_hungarian.isnull().sum())
# print removed missing values cleveland dataset df
removed_missing_counts1 = dfHeart_cleveland.isna().sum()
print("Cleveland Dataframe with no missing values:\n", removed_missing_counts1)

# print removed missing values hungarian dataset df
removed_missing_counts2 = dfHeart_hungarian.isna().sum()
print("\nhungarian Dataframe with no missing values:\n", removed_missing_counts2)
#Correlation Matrix analysis and plotting Heat map for cleveland heart disease dataset 
# define column names
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
           'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
# assign column names to the dataframe
dfHeart_cleveland.columns = columns
# create correlation matrix
corr = dfHeart_cleveland.corr()
# plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Cleveland Heart Disease Dataset')
plt.show()
# analyze correlations with respect to target variable
target_corr = corr['target'].drop('target')
print('Correlations with respect to target variable:\n')
print(target_corr.sort_values(ascending=False))
#Correlation Matrix analysis and plotting Heat map for hungarian heart disease dataset
# define column names
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
           'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
# assign column names to the dataframe
dfHeart_hungarian.columns = columns

# create correlation matrix
corr = dfHeart_hungarian.corr()
# plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Hungarian Heart Disease Dataset')
plt.show()
# analyze correlations with respect to target variable
target_corr = corr['target'].drop('target')
print('Correlations with respect to target variable:\n')
print(target_corr.sort_values(ascending=False))
#Creating histogram plots for analysis of cleveland dataset
# Create subplots for each attribute
fig, axs = plt.subplots(ncols=len(columns), figsize=(20, 5), gridspec_kw={'wspace': 0.4, 'hspace': 0.8})
# Plot the distribution of each attribute
for i, col in enumerate(columns):
    sns.histplot(dfHeart_cleveland[col], ax=axs[i], kde=True, color='royalblue')
    axs[i].set_title(col, fontsize=12)
    axs[i].set_xlabel("")
    axs[i].set_ylabel("")
    axs[i].tick_params(axis='both', which='major', labelsize=10, pad=2)
    for tick in axs[i].get_xticklabels():
        tick.set_rotation(45)
# Set the overall title and adjust the layout
fig.suptitle('Distribution of Heart Disease Dataset Attributes of Cleveland', fontsize=16, y=1.05)
#plt.tight_layout()
# Show the plot
plt.subplots_adjust(wspace=0.5)
plt.show()
#Creating histogram plots for analysis of hungarian dataset
# Create subplots for each attribute
fig, axs = plt.subplots(ncols=len(columns), figsize=(20, 5), gridspec_kw={'wspace': 0.4, 'hspace': 0.8})
# Plot the distribution of each attribute
for i, col in enumerate(columns):
    sns.histplot(dfHeart_hungarian[col], ax=axs[i], kde=True, color='royalblue')
    axs[i].set_title(col, fontsize=12)
    axs[i].set_xlabel("")
    axs[i].set_ylabel("")
    axs[i].tick_params(axis='both', which='major', labelsize=10, pad=2)
    for tick in axs[i].get_xticklabels():
        tick.set_rotation(45)
# Set the overall title and adjust the layout
fig.suptitle('Distribution of Heart Disease Dataset Attributes of hungarian', fontsize=16, y=1.05)
#plt.tight_layout()
# Show the plot
plt.subplots_adjust(wspace=0.5)
plt.show()
#"Randomized Cross-Validation with Decision Trees using Gini and Entropy Impurity Measures"- For Cleveland Heart Disease data
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.tree import DecisionTreeClassifier

dfHeart_cleveland = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]
#According to the attribute information for the Heart Disease UCI dataset
# Convert categorical variables to dummy variables
# Define the categorical columns
#cat_cols = [1, 2, 5, 6, 8, 10, 11]#Therefore, the categorical columns are sex, cp, fbs, restecg, exang, slope, and thal
#num_cols = [0, 3, 4, 7, 9]# age,trestbps,chol,thalach,oldpeak
# Convert categorical variables to dummy variables
dfHeart_cleveland = pd.get_dummies(dfHeart_cleveland, columns=["sex", "cp", "fbs", "restecg","exang", "slope", "thal"])
# Scale the numerical variables
scaler = StandardScaler()
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca"]
dfHeart_cleveland[num_cols] = scaler.fit_transform(dfHeart_cleveland[num_cols])
# Print the first 5 rows of the processed dataset
print("the first 5 rows of the processed cleveland heart disease dataset:\n",dfHeart_cleveland.head())
# Split into X and y
# Based on the correlation matrix analysis, the columns with low correlation coefficients are 'thalach' 'fbs' and 'chol'
# Therefore, it is safe to drop these columns from the X matrix when splitting the data into X and y.
# Dropping 'thalach' 'fbs' and 'chol' columns as they have low correlation with the target variable 'target' and are safe to drop
X = dfHeart_cleveland.drop(columns=['target', 'fbs', 'chol', 'thalach'])
y = dfHeart_cleveland['target']
print("\nColumns in X : \n", X.columns)


# Set the number of splits for cross-validation
k = 5
# Define the cross-validation method
cv_method = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
# Define the parameter grid to search
param_grid = {'criterion': ['gini', 'entropy'],
              'max_depth': [X.shape[1] * 10]}
print(X.shape[1])
# Create an empty list to store the accuracies
accuracies = []
# Loop over the folds
for train_idx, test_idx in cv_method.split(X, y):
   
    # Split the data into training and testing sets
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]
    
    # Initialize a list to store the accuracy values for this fold
    fold_accuracies = []
    
    # Loop over the impurity measures
    for impurity_measure in ['gini', 'entropy']:
        
        # Create the decision tree classifier with the current impurity measure
        dt = DecisionTreeClassifier(criterion=impurity_measure)
        
        # Create the grid search object
        grid_search = GridSearchCV(dt, param_grid, cv=cv_method)
        # Fit the grid search object to the training data
        grid_search.fit(X_train, y_train)
        
        # Get the best estimator
        best_dt = grid_search.best_estimator_
        
        # Evaluate the model on the testing data and append the accuracy to the list
        acc = best_dt.score(X_test, y_test)
        fold_accuracies.append(acc)
    
    # Append the fold accuracies to the overall accuracies list
    accuracies.extend(fold_accuracies)

# Print the accuracies
print("\nAccuracies:", accuracies)
print("Mean accuracy:", sum(accuracies)/len(accuracies))

# Separate the accuracies for each impurity measure
gini_accuracies = accuracies[:5]
entropy_accuracies = accuracies[5:]

# Compute the overall accuracy for Gini and Entropy
gini_mean_accuracy = np.mean(gini_accuracies)
entropy_mean_accuracy = np.mean(entropy_accuracies)
# Print the overall accuracies for Gini and Entropy
print("Overall accuracy for Gini:", gini_mean_accuracy)
print("Overall accuracy for Entropy:", entropy_mean_accuracy)
# Check which parameter gives the best results
best_param = grid_search.best_params_
print("Best parameter:", best_param)
# Determine which parameter gives the best results
if gini_mean_accuracy > entropy_mean_accuracy:
    print("Gini gives the best results")
else:
    print("Entropy gives the best results")

#"Randomized Cross-Validation with Decision Trees using Gini and Entropy Impurity Measures"- For Hungarian Heart Disease data
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.tree import DecisionTreeClassifier

dfHeart_hungarian = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]
#According to the attribute information for the Heart Disease UCI dataset
# Convert categorical variables to dummy variables
# Define the categorical columns
#cat_cols = [1, 2, 5, 6, 8, 10, 11]#Therefore, the categorical columns are sex, cp, fbs, restecg, exang, slope, and thal
#num_cols = [0, 3, 4, 7, 9]# age,trestbps,chol,thalach,oldpeak
# Convert categorical variables to dummy variables
dfHeart_hungarian = pd.get_dummies(dfHeart_hungarian, columns=["sex", "cp", "fbs", "restecg","exang", "slope", "thal"])
# Scale the numerical variables
scaler = StandardScaler()
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak","ca"]
dfHeart_hungarian[num_cols] = scaler.fit_transform(dfHeart_hungarian[num_cols])
# Print the first 5 rows of the processed dataset
print("the first 5 rows of the processed Hungarian heart disease dataset:\n",dfHeart_hungarian.head())
# Split into X and y
# Based on the correlation matrix analysis, the columns with low correlation coefficients are 'thalach' 'fbs' and 'chol'
# Therefore, it is safe to drop these columns from the X matrix when splitting the data into X and y.
# Dropping 'thalach' 'ca' and 'restecg' columns as they have very low correlation with the target variable 'target' and are safe to drop
X = dfHeart_hungarian.drop(columns=['target', 'ca', 'thalach','restecg'])
y = dfHeart_hungarian['target']
print("\nColumns in X : \n", X.columns)
# Set the number of splits for cross-validation
k = 5
# Define the cross-validation method
cv_method = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
# Define the parameter grid to search
param_grid = {'criterion': ['gini', 'entropy'],
              'max_depth': [X.shape[1] * 10]}
#print(X.shape[1])
# Create an empty list to store the accuracies
accuracies = []
# Loop over the folds
for train_idx, test_idx in cv_method.split(X, y):
    
    # Split the data into training and testing sets
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]
    
    # Initialize a list to store the accuracy values for this fold
    fold_accuracies = []
    
    # Loop over the impurity measures
    for impurity_measure in ['gini', 'entropy']:
        
        # Create the decision tree classifier with the current impurity measure
        dt = DecisionTreeClassifier(criterion=impurity_measure)
        
        # Create the grid search object
        grid_search = GridSearchCV(dt, param_grid, cv=cv_method)
        
        # Fit the grid search object to the training data
        grid_search.fit(X_train, y_train)
        
        # Get the best estimator
        best_dt = grid_search.best_estimator_
        
        # Evaluate the model on the testing data and append the accuracy to the list
        acc = best_dt.score(X_test, y_test)
        fold_accuracies.append(acc)
    
    # Append the fold accuracies to the overall accuracies list
    accuracies.extend(fold_accuracies)

# Print the accuracies
print("\nAccuracies:", accuracies)
print("Mean accuracy:", sum(accuracies)/len(accuracies))

# Separate the accuracies for each impurity measure
gini_accuracies = accuracies[:5]
entropy_accuracies = accuracies[5:]
# Compute the overall accuracy for Gini and Entropy
gini_mean_accuracy = np.mean(gini_accuracies)
entropy_mean_accuracy = np.mean(entropy_accuracies)
# Print the overall accuracies for Gini and Entropy
print("Overall accuracy for Gini:", gini_mean_accuracy)
print("Overall accuracy for Entropy:", entropy_mean_accuracy)
# Check which parameter gives the best results
best_param = grid_search.best_params_
print("Best parameter:", best_param)
# Determine which parameter gives the best results
if gini_mean_accuracy > entropy_mean_accuracy:
    print("Gini gives the best results")
else:
    print("Entropy gives the best results")
```
